## Portfolio Optimization

In this mini project I'll write a simple portfolio optimizer using the Sharpe Ratio evaluation metric.


```python
import numpy as np
import pandas as pd
import scipy.optimize as spo
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def get_data(start='2017-01-01', end='2017-12-31', symbols=['SPY','GOOG','GLD','IBM']):
    dates = pd.date_range(start, end)
    df = pd.DataFrame(index=dates)
    
    if 'SPY' in symbols:
        ref='SPY'
    else:
        ref = symbols[0]
        
    for s in symbols:
        temp = pd.read_csv("../../data/"+s+".csv",
                            index_col='Date',
                            parse_dates=True,
                            usecols=['Date','Adj Close'],
                            na_values=['nan'])
        temp = temp.rename(columns={'Adj Close':s})
        
        if s == ref:
            df = df.join(temp, how='inner') #use ref stock as reference
        else:
            df = df.join(temp) #default how='left'
    return df

def normalize(df):
    df = df/df.iloc[0]
    return df

def plot_data(dfs, title="Stock Data"):
    for df in dfs:
        df.plot(title=title)
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.show()
```

## The Portfolio

The test portfolio I'll use starts with a value of $1,000,000 and consists of the following stock allocation:
- S&P 500 Index - SPY - 40%
- Google - GOOG - 40%
- Gold - GLD - 10%
- IBM - IBM - 10%

The period being analyzed is a 3 year period from January 1, 2015 to December 31, 2017



```python
assets = ['SPY','GOOG','GLD','IBM']
alloc = [0.4, 0.1, 0.1, 0.4]
start_val = 1000000
start_date = '2015-01-01'
end_date = '2017-12-31'
```

### Data and Graphs
Now let's take a look at the data. The following are shown in order:
- Preview of Table of Adjusted Close Prices for each stock for each date
- Graph of Adjusted Close Prices for each stock
- Normalized Graph of adjusted close price for each stock
- Normalized Graph Scaled by percentage allocation of the stocks within the portfolio
- Graph of Actual Value of each holding within the portfolio
- Graph of Total Portfolio Value over Time

```python
df = get_data(start_date, end_date, assets)
df.head()
```

```python
plot_data([df])
```

```python
norm = normalize(df)
plot_data([norm], title="Normalized Price Data")
```

```python
alloced = norm*alloc
plot_data([alloced], title='Allocation Adjusted Price Data')
```

```python
values = alloced*start_val
plot_data([values], title='Value of Each Asset over Time')
```

```python
port_val = values.sum(axis=1)
print(port_val.head())
plot_data([port_val], "Total Portfolio Value over Time")
```

```python
def calc_port_val(alloc):
    values = norm*alloc*start_val
    port_val = values.sum(axis=1)
    return port_val
```

## Evaluation Metrics
There are many ways to evaluate a portfolio. Here are 4 of the most popular ways which I will cover:
- Cumulative Returns
- Average Daily Returns
- Standard Deviation of Daily Returns (measures volatility)
- Sharpe Ratio

### Cumulative Returns

```python
def calc_cum_ret(alloc):
    port_val = calc_port_val(alloc)
    cum_ret = port_val/port_val[0] - 1
    return cum_ret

cum_ret = calc_cum_ret(alloc)
print('Cumulative Returns:\n\n{}\n...'.format(cum_ret.head()))
plot_data([cum_ret], title='Cumulative Portfolio Returns')
print("Return for entire 3 year period:",cum_ret[-1])
```

### Daily Returns

```python
def calc_daily_ret(alloc):
    port_val = calc_port_val(alloc)
    daily_ret = port_val/port_val.shift(1) - 1
    daily_ret = daily_ret[1:]
    avg_daily_ret = daily_ret.mean()
    std_daily_ret = daily_ret.std()

    return daily_ret, avg_daily_ret, std_daily_ret

daily_ret, avg_daily_ret, std_daily_ret = calc_daily_ret(alloc)
print("Daily Returns:\n")
print('{}\n...'.format(daily_ret.head()))
print('Average Daily Returns: {}\nStandard Deviation of Daily Returns: {}'.format(avg_daily_ret, std_daily_ret))

plot_data([daily_ret],title='Daily Returns')

daily_ret.hist(bins=50)
plt.title('Histogram of Daily Returns')
plt.xlabel('Return')
plt.ylabel('Frequency')

plt.axvline(avg_daily_ret, color='g', linestyle='--', label='mean')
plt.axvline(std_daily_ret, color='r', linestyle='--',label='standard dev')
plt.axvline(-std_daily_ret, color = 'r', linestyle='--')
plt.legend()
plt.show()
```

### Sharpe Ratio
The shape ratio can be thought of as a risk adjusted return. 

The risk is measured in terms of how volatile the asset is. And the standard deviation of the daily returns of an asset is a good measure of the volatility of the asset. 
Therefore, the sharpe ratio divides the average daily return by the standard deviation

```python
def calc_sharpe(alloc):
    daily_ret, avg_daily_ret, std_daily_ret = calc_daily_ret(alloc) 
    
    sharpe = (avg_daily_ret)/std_daily_ret    
    return sharpe
    
sharpe = calc_sharpe(alloc)
print("Sharpe Ratio:",sharpe)
```

## Portfolio Optimization
Now that we have metrics by which to evaluate the portfolio, we can optimize the portfolio allocation by maximizing one of those metrics. 

The most trivial metric to maximize is the cumulative return. However, this would simply select the stock that has the highest return and eliminate the other assets from the portfolio. Therefore this isn't a good evaluation metric to optimize. 

We could optimize based on average daily returns. But notice that the Sharpe ratio is also proportional to average daily returns. In addition though, the Sharpe Ratio is inversely proportional to the volatility (risk) of the asset. Therefore, I think the Sharpe ratio is the best metric on which to evaluate the stock. 

Another good optimization metric could be a weighted combination of the sharpe ratio with the cumulative returns. But for this exercise I'll just do the sharpe ratio.

```python
def evaluate(X):
    sharpe = calc_sharpe(X)
    #cum_ret = calc_cum_ret(X)[-1]
    #return -(sharpe*0.99 + cum_ret*0.01) #alternate optimization metric metric
    return -sharpe #we will be maximizing sharpe, hence minimizing -sharpe
```

```python
best = spo.minimize(evaluate, [0.25,0.25,0.25,0.25], 
                    method='SLSQP',
                    options={'disp':True}, 
                    bounds=[(0.0,1.0)]*len(alloc), 
                    constraints={'type':'eq', 'fun': lambda x: x.sum() - 1})
best                  
```

## Comparison
Now let's compare the perfromance of the new asset allocation to the old one

```python
optimized_alloc = best.x
new_cum_ret = calc_cum_ret(optimized_alloc)
new_sharpe = calc_sharpe(optimized_alloc)

old_cum_ret = calc_cum_ret(alloc)
old_sharpe = calc_sharpe(alloc)
```

```python
print("Symbols: {}\nOld Allocation: {}\nNew Allocation: {}".format(assets, alloc, np.round(optimized_alloc,2)))

old_cum_ret.plot(label='Old Allocation')
new_cum_ret.plot(label='Optimized Allocation')
plt.title('Cumulative Portfolio Returns')
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.show()

print("Old Return for entire 3 year period:",old_cum_ret[-1])
print("Old Sharpe Ratio: ", old_sharpe,"\n")
print("New Return for entire 3 year period:",new_cum_ret[-1])
print("New Sharpe Ratio: ",new_sharpe)
```

## Conclusion
We see that the optimized portfolio allocation does far better than the one we had before. 

#### Caveat
It is important to note that this is a "hindsight" evaluation. We used past performance of a portfolio to determine which allocation would have been better. This is not to say that this portfolio is optimal for the future. Past performance does not indicate future performance.
