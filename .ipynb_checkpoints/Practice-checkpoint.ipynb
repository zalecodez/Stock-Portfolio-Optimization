```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
csvs = ['GOOG','SPY','IBM','GLD','XOM']
def getData(start, end, symbols=csvs):
    dates = pd.date_range(start, end)
    df = pd.DataFrame(index=dates)
    for f in symbols:
        temp = pd.read_csv("../data/"+f+".csv", 
                          index_col="Date",
                          parse_dates=True,
                          usecols=['Date','Adj Close'],
                          na_values=['nan'])
        temp = temp.rename(columns={'Adj Close':f})
        if f == 'SPY':
            df = df.join(temp,how='inner')
        else:
            df = df.join(temp)
    return df
```

```python
startdate = '2018-01-01'
enddate = '2018-05-01'

df = getData(startdate, enddate)
df
```

```python
df.plot()
plt.show()
```

```python
def normalize(df):
    df = df/df.ix[0]
    return df

def plot_data(dfs, title="Stock Data"):
    for df in dfs:
        df.plot()
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.show()
```

```python
plot_data([normalize(df)])
```

```python
roll_spy = df['SPY'].rolling(center=False,window=20)
rm_spy = roll_spy.mean()
rs_spy = roll_spy.std()

above = rm_spy + 2*rs_spy
below = rm_spy - 2*rs_spy

plot_data([df['SPY'],rm_spy,above,below], "Bollinger Bands (R)")
plt.show()
```

```python
def daily_returns(df):
    d = df.copy()
    d.ix[0] = 0
    d.ix[1:] = (df.ix[1:]/df.ix[:-1].values) - 1
    return d
```

```python
plot_data([daily_returns(df[['SPY','XOM']])])
```

```python
def cumulative_returns(df):
    return df/df[0] - 1
```

```python
plot_data([cumulative_returns(df['SPY'])])
```

### Dealing with Missing Data
1. Fill Forward first
2. Then Fill Backward

In python, use [pandas.DataFrame.fillna()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.fillna.html) to fill.

Fill forward: `df.fillna(method='pad')` or `df.fillna(method='ffill')`

Fill backward: `df.fillna(method='backfill')` or `df.fillna(method='bfill')`


```python
start = '2015-05-15'
end = '2018-05-15'

d10 = getData(start,end,['SPY','XOM','GLD'])
spy10 = d10['SPY']
xom10 = d10['XOM']
```

```python
plot_data([spy10])
```

```python
daily = daily_returns(spy10)
daily.hist(bins=50)
plt.show()
```

```python
mean = daily.mean()
std = daily.std()
kurtosis = daily.kurtosis()
print("mean =",mean)
print("std =",std)
print("kurtosis =",kurtosis)
```

```python
daily.hist(bins=20)
plt.axvline(mean,color='w',linestyle='dashed',linewidth=2)
plt.axvline(std, color='r',linestyle='dashed',linewidth=2)
plt.axvline(-std, color='r',linestyle='dashed',linewidth=2)
plt.show()
```

Higher return corresponds to higher mean 
Higher volatility corresponds to higher standard deviation

```python
daily = daily_returns(d10)
plot_data([daily])
```

```python
daily['SPY'].hist(bins=20,label='SPY')
daily['XOM'].hist(bins=20,label='XOM')
plt.show()
```

```python
plt.scatter(daily['SPY'],daily['XOM'])
beta,alpha = np.polyfit(daily['SPY'],daily['XOM'],1)
plt.plot(daily['SPY'], beta*daily['SPY']+alpha,'-',color='r')

plt.show()
```
